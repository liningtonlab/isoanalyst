```python
import pandas as pd
from utils import *
conditions = ['ACE','PROP','MET','GLU']
files = getfilenames_cppis('./IsotopeDataProcessing/TEST/CPPIS/', conditions)
```

```python
files.head()
```

```python
reps = files[files['condition']=='BLANK']
dfs = []
for f in reps['path']:
    df = pd.read_csv(f)
    drop_cppis(df, inplace=True)
    dfs.append(df)
df = combine_dfs(dfs)    
group_cppis(df, 'BLANK', 'reps')
```

```python
df.reps.unique().shape
```

```python
# Remove data which doesn't meet min_reps
col='reps'
grouped = df.groupby(col)
df.shape
```

```python
min_reps = 3
to_drop = []
for idx, g in grouped:
    if g.shape[0] < min_reps:
        to_drop.extend(g.index)
df.drop(to_drop, inplace=True)
```

```python
df.head()
```

```python
# Average leftover data
# df.reset_index(inplace=True)
grouped = df.groupby(col)
new_data =[]
for idx, grp in grouped:
    new_data.append(
    {
        "Sample": grp.Sample.iloc[0].split('_')[0],
        "PrecMz": round(grp.PrecMz.mean(), 3),
        "PrecZ": grp.PrecZ.iloc[0], 
        "PrecIntensity": np.int64(grp.PrecIntensity.mean()),
        "LowScan": grp.ScanLowRange.min(),
        "HighScan": grp.ScanHighRange.max(),
        "reps": idx
    })
pd.DataFrame(new_data).head()
```

```python
# blank subtraction
blanks = pd.read_csv("IsotopeDataProcessing/TEST/BLANKS/All_ions_grouped.csv")
primary = pd.read_csv("IsotopeDataProcessing/TEST/ACE/All_ions_averaged.csv")
```

```python
drop_me = set()
df = primary
for data in blanks.itertuples():
    mz_range = ppm_tolerance(data.PrecMz)
    rt_range = rt_tolerance(data.RetTime)
    low_mz = df['PrecMz'] > mz_range[0]
    high_mz = df['PrecMz'] < mz_range[1]
    low_rt = df['RetTime'] > rt_range[0]
    high_rt = df['RetTime'] < rt_range[1]
    matching_z = df['PrecZ'] == data.PrecZ

    droppable = df[low_rt&high_rt&low_mz&high_mz&matching_z].index
    drop_me.update(droppable)
```
